```python
def limpeza_de_texto(tweet):
  tweet = tweet.lower()
  tweet = re.sub('\n', '', tweet)
  tweet = re.sub('@[^\s]+', '', tweet)
  tweet = re.sub('\B#\S+', '', tweet)
  tweet = re.sub('https?://\S+|www\.\S+', '', tweet)
  return tweet
```

```python
dataframe_final_candidatos['tweet_limpo'] = dataframe_final_candidatos['Tweet'].apply(lambda i:limpeza_de_texto(i))
```
